```python
import requests
import json
import pandas as pd
```

```python
with open('brewDB_key.txt', 'r') as file:
    for line in file:
        api_key = line
```

```python
url = r"https://sandbox-api.brewerydb.com/v2/breweries/?key="
```

```python
response = requests.get(url+api_key)
```

```python
idls = [] #ID of the beer
namels = [] #name of the beer
shortnamels = [] #name of the beer
descriptionls = [] #display name of the beer
websitels = [] #alcohol by volume
establishedls = [] #gravity of beer
organicls = [] #IBU (internationall bittering unit) - how bitter a beer is
imagesls = [] #temperature a beer should be served at
statusls = [] #the type of glass it should be served in
status_displayls = [] #Y/N if it is organic
createdatels = [] #Y/N if the beer is retired
updatedatels = [] #description of the beer
massownedls = [] #name of the style of beer
inbusinessls = [] #description of the style of beer
verifiedls = [] #date the beer was put in the system
```

```python
for brew in brewery:
        
    #get the value of each key and if the key doesn't exist, set a variable to be None
    try: bid = brew['id']
    except: bid = None
            
    try: name = brew['name']
    except: name = None
            
    try: sname = brew['nameShortDisplay']
    except: sname = None
            
    try: description = brew['description']
    except: description = None
            
    try: website = brew['website']
    except: website = None
            
    try: established = brew['established']
    except: ibu = None
            
    try: organic = brew['isOrganic']
    except: organic = None
            
    try: images = brew['images']
    except: images = None
            
    try: status = brew['status']
    except: status = None
            
    try: status_display = brew['statusDisplay']
    except: status_display = None
            
    try: createdate = brew['createDate']
    except: createdate = None
            
    try: updatedate = brew['updateDate']
    except: updatedate = None
            
    try: massowned = brew['isMassOwned']
    except: massowned = None
            
    try: inbusiness = brew['isInBusiness']
    except: inbusiness = None
            
    try: verified = brew['isVerified']
    except: verified = None
            
        
    idls.append(bid)
    namels.append(name)
    shortnamels.append(sname)
    descriptionls.append(description)
    websitels.append(website)
    establishedls.append(established)
    organicls.append(organic)
    imagesls.append(images)
    statusls.append(status)
    status_displayls.append(status_display)
    createdatels.append(createdate)
    updatedatels.append(updatedate)
    massownedls.append(massowned) 
    inbusinessls.append(inbusiness)
    verifiedls.append(verified)
```

```python
breweryList = list(zip(idls, namels, shortnamels, 
                    descriptionls, websitels, establishedls, organicls, 
                    imagesls, statusls, status_displayls, createdatels, 
                    updatedatels, massownedls, inbusinessls, verifiedls))

#names for columns in data frame
colnames = ['id', 'name', 'short_name', 'description', 'website', 'established', 
            'organic', 'images', 'status', 'status_display', 'createdate', 
            'updatedate', 'massowned', 'inbusiness', 'verified']
```
