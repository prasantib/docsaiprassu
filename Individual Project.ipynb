```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

location = "heart.csv"
df = pd.read_csv(location)
df
```

```python
df.dtypes
```

```python
#printing the means for various attributes that are considered to be attributed to cardiovascular disease
print(df['age'].mean())
print(df['cp'].mean())
print(df['chol'].mean())
print(df['thalach'].mean())
print(df['oldpeak'].mean())
```

```python
#printing the modes for each attribute
print(df['age'].mode())
print(df['cp'].mode())
print(df['chol'].mode())
print(df['thalach'].mode())
print(df['oldpeak'].mode())
```

```python
#density plot in order to look at various variables
plt.figure(figsize=(10,6))

ax = sns.kdeplot(df['age'][df['cp'] == 1], 
                 color="darkturquoise",
                 shade=True)

sns.kdeplot(df['age'][df['target'] == 0], 
            color="lightcoral",
            shade=True)

plt.legend (['Chest Pain', 'Target'])
plt.title("Density Plot of Age for Chest Pain vs Presence of Disease")

ax.set(xlabel='Age')
#plt.show()
```

```python
#get the number of missing values in each column
df.isnull().sum()
```

```python
#grouping by sex, chest pain, and if the patient has heart disease, by the average age.
#This shows the gender, and what was the severity of their chest pain, and if they had heart disease. 
df.groupby(['sex', 'cp', 'target'])['age'].mean()
```

```python
#grouping by major indicators of cardiovascular disease
df.groupby(['age', 'cp', 'chol'])['age'].mean()
```
